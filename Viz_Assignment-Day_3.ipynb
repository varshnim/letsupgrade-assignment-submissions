Plot using sns.relplot, line plot, box plot on fmri data set

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from numpy.random import randn
import seaborn as sns
```

```python
fmri_data = sns.load_dataset('fmri')
fmri_data = fmri_data.sort_values(by=['subject','timepoint'])
```

```python
fmri_data
```

## Relplot

```python
sns.relplot(
    data=fmri_data, 
    x='subject', 
    y = 'signal',
    col = 'event',
    size = 'timepoint',
    
)

plt.show()
```

## Lineplot

```python
sns.lineplot(
    data = fmri_data,
    x = 'timepoint',
    y = 'signal',
)
plt.title('Seaborn Lineplot')
plt.show()
```

## Boxplot

```python
sns.color_palette("mako")
sns.boxplot(
    data = fmri_data,
    x = 'subject',
    y = 'signal',
    hue = 'event',
    palette = 'mako'
)
plt.title('Seaborn boxplot')
plt.show()
```
