# Question 2
Plot a line plot showing the sales trend in company 1 and 2:
<br>
days = [1,2,3,4,5,6,7] #days of d week<br>
sales_1 = [160,150,140,145,175,165,180] #sales of company1<br>
sales_2 = [70,90,160,150,140,145,175]  #sales of company2


```python
days = [1,2,3,4,5,6,7] #days of d week
sales_1 = [160,150,140,145,175,165,180] #sales of company1
sales_2 = [70,90,160,150,140,145,175]  #sales of company2
```

```python
plt.plot(days, sales_1,label="Company_1")
plt.plot(days, sales_2,label="Company_2")
plt.legend()
plt.xlabel('Day of Week')
plt.ylabel('Daily Sales')
plt.title('Comparison of Sales Trends of Company 1 & Company 2')
plt.show()
```

# Question 3
Create a 3 by 3 subplots: <br>
multiple plots<br>
x = [1,2,3,4]<br>
y1 = [4,3,2,1]<br>
y2 = [10,20,30,40]<br>
y3 = [40,30,20,10]<br>
y4 = [1,2,1,2]<br>
y5 = [40,70,90,70]


```python
x = [1,2,3,4]
y1 = [4,3,2,1]
y2 = [10,20,30,40]
y3 = [40,30,20,10]
y4 = [1,2,1,2]
y5 = [40,70,90,70]
```

```python
plt.figure(figsize=(15,10))

plt.subplot(3,3,1)
plt.plot(x,y1,"r")
plt.title("x & y1",y=-0.2)

plt.subplot(3,3,2)
plt.plot(x,y2,"go-")
plt.title("x & y2",y=-0.2)

plt.subplot(3,3,3)
plt.plot(x,y3,"r")
plt.title("x & y3",y=-0.2)

plt.subplot(3,3,4)
plt.plot(x,y4,"bo--")
plt.title("x & y4",y=-0.2)

plt.subplot(3,3,5)
plt.plot(x,y5,"r--")
plt.title("x & y5",y=-0.2)

plt.subplot(3,3,6)
plt.plot(y1,y2,"r")
plt.title("y1 & y2",y=-0.2)

plt.subplot(3,3,7)
plt.plot(y1,y3,"go-")
plt.title("y1 & y3",y=-0.2)

plt.subplot(3,3,8)
plt.plot(y1,y4,"bo--")
plt.title("y1 & y4",y=-0.2)

plt.subplot(3,3,9)
plt.plot(y1,y5,"go-")
plt.title("y1 & y5",y=-0.2)

plt.suptitle("Subplots")
plt.show()
```
